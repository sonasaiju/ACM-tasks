```python
# Import pandas for data handling and seaborn/matplotlib for data visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv") # Load the Students Performance dataset
df.head() # Display the first five rows
```

```python
df.info() # Get a concise summary of the dataset
```

```python
df.columns # Display all column names
```

```python
df.isnull().sum() # to find all missing errors
```

```python
# Visualize the distribution of math scores to understand overall performance and score spread

sns.histplot(df["math score"], kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

Math scores follow an approximately normal distribution, with most students scoring between 50 and 80. The peak around 6070 shows the most common performance range, and only a few students score very low or very high.

```python
# Create a scatter plot to examine the relationship between reading and writing scores
sns.scatterplot(x="reading score", y="writing score", data=df)
plt.title("Reading Score vs Writing Score")
plt.show()
```

There is a strong positive correlation between reading and writing scores,students who score high in reading tend to score high in writing. This suggests these two skills are closely related and improve together.

```python
# Compare math score distributions across genders using a box plot
sns.boxplot(x="gender", y="math score", data=df)
plt.title("Math Score by Gender")
plt.show()
```

The boxplot compares math score distributions between male and female students. Both genders show similar score ranges, with males having a slightly higher median. Overall, the difference in math performance by gender is small, indicating comparable achievement levels.

```python
# Compare math scores of students based on completion of the test preparation course
sns.boxplot(x="test preparation course", y="math score", data=df)
plt.title("Math Score vs Test Preparation Course")
plt.show()
```

Students who completed the test preparation course have a higher median math score compared to those who did not. The score distribution for the completed group is shifted upward, indicating a positive impact of test preparation on math performance.

```python
# Analyze differences in math scores across lunch categories to explore socio-economic patterns
sns.boxplot(x="lunch", y="math score", data=df)
plt.title("Math Score by Lunch Type")
plt.show()
```

Students with a standard lunch generally score higher in math than those receiving free/reduced lunch. This suggests a possible association between lunch type and academic performance.

```python
# Compute correlation between math, reading, and writing scores to quantify their relationships
df[["math score", "reading score", "writing score"]].corr()
```

There is a strong positive correlation between reading and writing scores, indicating students who perform well in reading also tend to do well in writing. Math scores also show a strong positive relationship with both reading and writing, suggesting overall academic performance is closely related across subjects.

```python
# Calculate skewness of math scores to understand the symmetry of the score distribution
df["math score"].skew()
```

The skewness value is slightly negative, indicating the math score distribution is mildly left-skewed. This means a small number of lower scores pull the distribution to the left, while most students score around the middle to higher range.

Analysis

The data shows strong positive correlations between reading, writing, and math scores, indicating that students who perform well in one subject tend to perform well in others. Clear patterns are observed where test preparation and standard lunch are associated with higher math scores. A few low-score outliers appear across groups, but no extreme or inconsistent anomalies are present.No missing values or inconsistent data types were found in the dataset, indicating good overall data quality.
